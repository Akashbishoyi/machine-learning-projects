# 1. Handling the missing values

```python
import pandas as pd
```

```python
df=pd.read_csv("ipl2017.csv")
```

```python
df.info()
```

```python

df.head()
```

```python
df.shape
```

## 2.Drop the unnecessary columns

```python
df=df.drop('mid',axis=1)
```

```python
df=df.drop('date',axis=1)
```

```python
df.info()
```

```python
df.shape
```

```python
y=df["total"]
```

```python
X=df.drop('total',axis=1)
```

```python
df.info()
```

```python
type(X)
```

```python
X.shape
```

## 3.Convert the categorical string columns to numerical columns by using one-hot encoding

```python
X=pd.get_dummies(X,columns=["venue"])
```

```python
X=pd.get_dummies(X,columns=["bat_team"])
```

```python
X=pd.get_dummies(X,columns=["bowl_team"])
```

```python
X=pd.get_dummies(X,columns=["batsman"])
```

```python
X=pd.get_dummies(X,columns=["bowler"])
```

```python
X.shape
```

## Train-Test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)
```

# Perform feature scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

## 5. Build a model on the "Total" column,using a RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forest=RandomForestRegressor()
```

```python
forest.fit(X_train,y_train)
```

## 6. Calculating the score

```python
forest.score(X_test,y_test)
```

```python
forest.score(X_train,y_train)
```
